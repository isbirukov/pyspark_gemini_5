### The Simple Path
- Создайте DataFrame из 10 чисел.
- Умножьте каждое на 2.
- Выведите результат.

Вопрос: Сколько Jobs и Stages было создано? (Ответ: 1 Job, 1 Stage).

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder \
        .appName("Simple Path") \
        .master("local[*]") \
        .getOrCreate()

df = spark.range(0, 10).withColumn(
    "id",
    col("id") * 2
)
df.show()
```

```python
spark.stop()
```

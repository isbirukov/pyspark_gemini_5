### Double Shuffle (Advanced)
- Создайте данные.
- Сгруппируйте их по остатку от деления на 2 (groupBy).
- Посчитайте количество.
- Сделайте repartition(1) результата.
- Сохраните в фиктивный формат (или вызовите collect).

Цель: Увидеть цепочку Stages.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, expr, count

spark = SparkSession.builder \
        .appName("Double Shuffle") \
        .master("local[*]") \
        .getOrCreate()

df = spark.range(0, 1000000).groupBy(expr("id % 2")).agg(
    count("*")
)
df.printSchema()
df.show(10)
repartitioned_data = df.repartition(1)
repartitioned_data.collect()
```

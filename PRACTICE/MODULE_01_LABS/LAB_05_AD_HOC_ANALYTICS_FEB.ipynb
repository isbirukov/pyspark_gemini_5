## Создать, вывести все конфигурации (getConf().getAll()).

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
        .appName("AD_HOC_ANALYTICS_FEB") \
        .master("local[*]") \
        .getOrCreate()

print(*spark.sparkContext.getConf().getAll(), sep=("\n"))
spark.stop()
```

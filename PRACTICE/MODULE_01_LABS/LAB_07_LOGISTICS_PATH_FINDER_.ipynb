## Создать, дважды вызвать .getOrCreate(). Проверить имя.

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
        .appName("LOGISTICS_PATH_FINDER") \
        .master("local[*]") \
        .getOrCreate()

spark = SparkSession.builder \
        .appName("LOGISTICS_PATH_FINDER_2") \
        .master("local[*]") \
        .getOrCreate()



print(spark.sparkContext.appName)
spark.stop()
```

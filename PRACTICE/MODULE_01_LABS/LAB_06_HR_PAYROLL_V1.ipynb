## Создать сессию, используя .config("spark.app.name", "HR_PAYROLL_V1") вместо .appName().

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
        .config("spark.app.name", "HR_PAYROLL_V1") \
        .master("local[*]") \
        .getOrCreate()

print(spark.sparkContext.appName)
spark.stop()
```

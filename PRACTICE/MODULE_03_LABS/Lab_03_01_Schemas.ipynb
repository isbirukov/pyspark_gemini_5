```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType, DateType

spark = SparkSession.builder.appName("DataFrame_Basics").getOrCreate()
```

Попробуйте создать DataFrame с inferSchema (просто передав список). Посмотрите, какой тип Spark даст второму полю (скорее всего, String, чтобы вместить "TWENTY")

```python
bad_data = [
    ("Alice", 34, "2023-01-01"),
    ("Bob", "TWENTY", "2023-05-05"), # Ошибка типа в возрасте
    ("Charlie", 45, None)
]

df_corrupted = spark.createDataFrame(bad_data)
df_corrupted.printSchema()
```

Создайте жесткую схему: Name (String), Age (Integer), JoinDate (Date).
Попробуйте создать DataFrame с этой схемой. Ожидаемый результат: Spark 3.x может выбросить ошибку или превратить "TWENTY" в null (зависит от настроек парсера, но при создании из списка Python он упадет с TypeError).

Цель: Убедиться, что схема защищает качество данных.

```python
data_schema = StructType([
    StructField("Name", StringType(), True),
    StructField("Age", IntegerType(), True),
    StructField("JoinDate", DateType(), True)
])

df_schema = spark.createDataFrame(bad_data, schema=data_schema)
df_schema.printSchema()
```
